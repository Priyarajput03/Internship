### Q 1Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank          B) Name           C) Artist        D) Upload date          E) Views


```python
driver=webdriver.Chrome(os.getcwd()+"\chromedriver.exe")
```

```python
#send get request to web page server for source code
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos '
page= requests.get(url)
#see page content
soup= BeautifulSoup(page.content)
driver.get(url)
driver.maximize_window()
```

```python
# fetaching details of all the rank contain in table column
Rank=[]
Name=[]
Artist=[]
Date=[]
Views=[]

try:
    for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]'):
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('---')
    
#fetching details of names in videos
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]"):
        Name.append(i.text)
except NoSuchElementException:
    Name.append('---')
    
#fetching Artist of the videos
try:
    for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]'):
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('---')
    
#fetching Date of the videos
try:
    for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]'):
        Date.append(i.text)
except NoSuchElementException:
    Date.append('---')
    
#fetching Views of the videos
try:
    for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]'):
        Views.append(i.text)
except NoSuchElementException:
    Views.append('---')
```

```python
df=pd.DataFrame({'Rank':Rank,'Name':Name,'Artist':Artist,'Date':Date,'Views':Views})

# Removing stray numbers from Name column
df.Name = df.Name.apply(lambda x:x[:-4].strip('"'))
df
```

```python
Name=[]
Data_type=[]
task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
```

```python
#Searching all the dataset name tags
name=driver.find_elements_by_xpath("//p[@class='normal']/b/a")[:623]

#scraping dataset name from tags
for i in name:
    Name.append(i.text.replace('\n',' '))

#finding all the datatype tags
datatype = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p')
#scraping datatype from tags
for i in datatype:
    Data_type.append(i.text.replace('\n',' '))

#finding all the task tags
Task = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p')
#scraping task from tags
for i in Task:
    task.append(i.text)
    
#finding all the attribute tags
attribute = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p')
#scraping attribute type from tags
for i in attribute:
    Attribute_type.append(i.text)
    
#finding all the instances tags
instance= driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p')

#scraping number of instances from tags
for i in instance:
    No_of_instances.append(i.text)
    
#finding all the number of attributes tags
Attribute=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p')
#scraping number of attributes from tags
for i in Attribute:
    No_of_attribute.append(i.text)
    
#finding all the number of attributes tags
year=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p')
#scraping year from tags
for i in year:
    Year.append(i.text)
```
