```python
Brand=[]
Model=[]
Variant=[]
Manufacturing_year=[]
Driven_kilometers=[]
Fuel=[]
No_of_owners=[]
Location=[]
Price=[]
```

```python
def Cars(location):
   
    model=driver.find_elements_by_xpath("//div[@class='_1l4fi']/div/div/h2")
    for i in model:
        #print(i.text)
        str=i.text
        m1=str.split(" " ,2)
        model=str.split()[1]
        Model.append(m1[2])
        Manufacturing_year.append(m1[0])
        Brand.append(m1[1])

    variant=driver.find_elements_by_xpath("//div[@class='_1l4fi']/div/div/p")
    for i in variant:
        #print(i.text)
        Variant.append(i.text)

    Km=driver.find_elements_by_xpath("//div[@class='_1l4fi']/ul/li[1]")
    for i in Km:
        #print(i.text)
        str1=i.text
        str1=str1.split()[0] 
        str1=str1.replace(",","")
        #print(str1)
        str1=int(str1)
        #print(type(str1))
        Driven_kilometers.append(str1)

    owner=driver.find_elements_by_xpath("//div[@class='_1l4fi']/ul/li[2]")
    for i in owner:
        #print(i.text)
        No_of_owners.append(i.text)

    fuel=driver.find_elements_by_xpath("//div[@class='_1l4fi']/ul/li[3]")
    for i in fuel:
        #print(i.text)
        Fuel.append(i.text) 
 

    price=driver.find_elements_by_xpath("//div[@class='_1l4fi']/div[2]/div[2]/span")
    for i in price:
        #print(i.text)
        str2=i.text
        str2=str2[1:]
        
        str2=str2.replace(",","")
        #print(str2)
        str1=int(str2)
        #print(type(str2))
    
        Price.append(str2) 
        Location.append(location)
```

```python
Cars('New Delhi')
```

```python
df=pd.DataFrame({})
df['Brand']=Brand
df['Model']=Model
df['Year']=Manufacturing_year
df['Variant']=Variant
df['Location']=Location
df['Version']=Fuel
df['Number of Owners']=No_of_owners
df['KmDriven']=Driven_kilometers
df['Price']=Price
```
