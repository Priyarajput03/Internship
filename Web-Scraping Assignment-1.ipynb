##### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

```python
imdb=requests.get('https://www.imdb.com/chart/top/?sort=rk,asc&mode=simple&page=1')
soup1=BeautifulSoup(imdb.content,'html.parser')

#Movies Name
movie_titles= soup1.find_all("td", class_="titleColumn")[:100]
movies_name = [] #empty list
for i in movie_titles:
    for j in i.find_all("a"):
        movies_name.append(j.text)
movies_name

# Movies Release Year
movies_year=soup1.find_all('span', class_="secondaryInfo")[:100]
Year=[]
for i in movies_year:
    Year.append(i.text)
Year

#Movie Ratings
movies_rating=soup1.find_all('td',class_="ratingColumn imdbRating")[:100]
Rating=[]
for i in movies_rating:
    Rating.append(i.text.replace('\n',''))
Rating

print(len(movies_name),len(Year),len(Rating))

import pandas as pd
df=pd.DataFrame({'Movies':movies_name,'Year':Year,'Rating':Rating})
df
```

##### 3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
#send get request to web page server for source code
url = "https://www.imdb.com/india/top-rated-indian-movies/"
page3 = requests.get(url)
#see page content
soup3 = BeautifulSoup(page3.content)

Indian_Movies=soup3.find_all('td', class_="titleColumn")  #top Indian movies name

Imdb_Indian_Movies=[]
for i in Indian_Movies:
    for j in i.find_all("a"):
        Imdb_Indian_Movies.append(j.text)
Imdb_Indian_Movies


Imdb_year=soup3.find_all('span', class_="secondaryInfo")  # Year of Release
Year3=[]
for i in Imdb_year:
    Year3.append(i.text)
Year3

Movies_rating3=soup3.find_all('td',class_="ratingColumn imdbRating") # Movie Ratings
Rating3=[]
for i in Movies_rating3:
    Rating3.append(i.text.replace('\n',''))
Rating3

print(len(Imdb_Indian_Movies),len(Year3),len(Rating3))

df3=pd.DataFrame({'Movies':Imdb_Indian_Movies,'Year':Year3,'Rating':Rating3})

df3.drop(df3.tail(150).index,inplace = True)
df3
```

##### 4.Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp . 

```python
#send get request to the webpage server to get the source code of the page
url = "https://meesho.com/bags-ladies/pl/p7vbp"
page4 = requests.get(url)

#see page content
soup4 = BeautifulSoup(page4.content)

# Bag's name
Bag_name = soup4.find_all('p', class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS") 
Bags = [] #empty list
for i in Bag_name:
    Bags.append(i.text)
Bags

# Bag's prices
Bag_prices = soup4.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy") 
price = [] #empty list
for i in Bag_prices:
    price.append(i.text.replace('₹',''))
price

 # Discount on Bag's
Bag_Discount = soup4.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 fCJVtz NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm") # Discount pecentage
Discount = [] #empty list
for i in Bag_Discount:
    Discount.append(i.text.replace('₹','').split()[0])
Discount
  
print(len(Bag_name),len(Bag_prices),len(Bag_Discount))

df4=pd.DataFrame({'Bag_name':Bags,'Bag_prices':price,'Bag_Discount':Discount})
df4
```

##### 5. Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:    
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.



```python
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
page5 = requests.get(url)
soup5 = BeautifulSoup(page5.content)

#Top 10-Men's Team
Team=soup5.find_all('span',class_="u-hide-phablet")[:10]
Top10_Teams=[]
for i in Team:
    Top10_Teams.append(i.text)
Top10_Teams

# Top 10 Matches
Matches1=soup5.find_all('td', class_="rankings-block__banner--matches")
ODI_Matches1=[]
for i in Matches1:
    ODI_Matches1.append(i.text)
ODI_Matches1

Matches2=soup5.find_all('td', class_="table-body__cell u-center-text")
ODI_Matches2=[]
for i in Matches2[:18:2]:
    ODI_Matches2.append(i.text)
ODI_Matches2

Top10_Matches=ODI_Matches1 + ODI_Matches2
Top10_Matches

#Top 10 Matches Points
Points1=soup5.find_all('td', class_="rankings-block__banner--points")
Points2=soup5.find_all('td', class_="table-body__cell u-center-text")

ODI_Points1=[]
for i in Points1:
    ODI_Points1.append(i.text)
ODI_Points1

ODI_Points2=[]
for i in Points2[1:18:2]:
    ODI_Points2.append(i.text)
ODI_Points2

Points=ODI_Points1 + ODI_Points2
Points

#Top10 Match ratings
Rating=soup5.find_all('td', class_="rankings-block__banner--rating u-text-right")
ODI_Rating5=[]
for i in Rating:
    ODI_Rating5.append(i.text.replace('\n','').strip())
ODI_Rating5  

Ratings5=soup5.find_all('td', class_="table-body__cell u-text-right rating")
ODI_Rating=[]
for i in Ratings5[:9]:
    ODI_Rating.append(i.text.replace('\n','').strip())
ODI_Rating

Rating=ODI_Rating5 +ODI_Rating
Rating
print(len(Team),len(Top10_Matches),len(Points),len(Rating))
df5=pd.DataFrame({'Team':Team, 'Matches':Top10_Matches,'Points':Points,'Rating':Rating})
df5
```
